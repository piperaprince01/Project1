```python
def Goal_diff_count(df1):
    #print(df1.head())
    tempdf = df1.copy(deep=True)
    tempdf.set_index('Team',inplace=True)
    #print(tempdf.head())
    GoalsFor = tempdf['GoalsFor'].astype(int)
    GoalsAgainst = tempdf['GoalsAgainst'].astype(int)
    
    Goal_diff_count = GoalsFor - GoalsAgainst
    #print(Goal_diff_count.head())
    return(Goal_diff_count,Goal_diff_count[Goal_diff_count==Goal_diff_count.max()],Goal_diff_count[Goal_diff_count==Goal_diff_count.min()])
    
```

```python
Goal_diff_count(df1)
```
